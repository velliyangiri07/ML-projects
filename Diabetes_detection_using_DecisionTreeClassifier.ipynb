```python
import pandas as pd
df = pd.read_csv("diabetes.csv")
df
```

```python
df.isnull().sum()
```

# separate the features and the target variable

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
x
```

```python
y
```

# splits the dataset into training and testing sets
1.  test_size=0.2
  * 20% of the data will be used for testing
  * 80% will be used for training
2.  random_state=0
  * get the same split every time

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

# imports the DecisionTreeClassifier from scikit-learn

```python
from  sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

# Train the model using Training data

```python
clf.fit(x_train,y_train)
```

# Make predictions on the test input data

```python
y_pred = clf.predict(x_test)
```

# calculates the accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# visualize your trained Decision Tree model


*   feature_names=x.columns: shows the actual column names instead of generic feature names like X[0], X[1], etc.

*   filled=True: colors the nodes based on class outcome (makes it easy to understand)



```python
from sklearn.tree import export_graphviz
import graphviz
graphviz.Source(export_graphviz(clf,feature_names=x.columns,filled = True))
```
