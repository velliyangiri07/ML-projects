```python
import pandas as pd
df = pd.read_excel('Restaurant.xlsx')
df.head()
```

```python
df.describe()
```

# average sales amount for October 1st, 2023

```python
filter = df[df['DATE']=='01.10.23']
res=filter['AMOUNT'].mean()
print(res)
```

# total bill value for invoice_no = 117

```python
invoice = df[df['INVOICE_NO'] == 117]
total_amount = invoice['AMOUNT'].sum()
print(total_amount)
```

# top 5 items with the highest total sales in the dataset.

```python
top5=df.groupby('ITEM')['SUB_TOTAL'].sum().sort_values(ascending=False).head(5)
print(top5)
```

```python
import matplotlib.pyplot as plt
plt.bar(top5.index,top5.values)
plt.show()
```

# top 5 most sold items

```python
top5=df.groupby('ITEM')['QUANTITY'].sum().sort_values(ascending=False).head(5)
top5
```

```python
plt.bar(top5.index,top5.values)
plt.show()
```

# top 1 highest-billed invoice

```python
df.groupby('INVOICE_NO')['AMOUNT'].sum().sort_values(ascending=False).head(1)
```

# total number of PAROTTA sold on 1st October 2023 between 6 PM and 9 PM

```python
a = df[
    (df['DATE'] == '01.10.23') &
    (df['ITEM'] == 'PAROTTA') &
    (df['TIME'] >= 18.00) &
    (df['TIME'] <= 21.00)
]

print(a['QUANTITY'].sum())
```
